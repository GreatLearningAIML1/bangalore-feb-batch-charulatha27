# 2. Question: Use the five percent significance test over the data to determine the p value to check new scheme has significantly raised outputs? (10 points)

```python
t_stat, p_val =  ttest_ind(df['Old Scheme (in thousands)'], df['New Scheme (in thousands)'])
print("five percent significance test, t_stat value is ", t_stat)
print("five percent significance test, P value is", p_val)
print("\n")

print("Null hypothesis is:  'There is no significant difference between Old and New Scheme'\n")
if p_val <= 0.05:
    print("P value is:", p_val, "\nSince p_val <= 0.05, we reject null hypothesis.")
    print("Hence, null hypothesis, that 'there is no significant difference between Old and New Scheme', is false.")
    print("Therefore new scheme has either decreased or raised outputs significantly.")
else:
    print("P value is:", p_val, "\nSince p_val > 0.05, we accept null hypothesis.")
    print("Hence, null hypothesis, 'that there is no significant difference between Old and New Scheme', is true.")
    print("Therefore new scheme has not raised outputs significantly.")
```

# 3. Question: What conclusion does the test (p-value) lead to? (2.5 points)

```python
print("Null hypothesis is:  'There is no significant difference between Old and New Scheme'\n")

print("p-value is", p_val)
      
if p_val <= 0.05:
    print("p-value is <= 0.05 and it indicates strong evidence against the null hypothesis, so we reject the null hypothesis.")
    print("Hence, null hypothesis, that 'there is no significant difference between Old and New Scheme', is false.")
    print("Therefore new scheme has either decreased or raised outputs significantly.")
else:
    print("p-value is > 0.05 and it indicates weak evidence against the null hypothesis, so we fail to reject the null hypothesis.")
    print("Hence, null hypothesis, 'that there is no significant difference between Old and New Scheme', is true.")
    print("Therefore new scheme has not raised outputs significantly.")
```
