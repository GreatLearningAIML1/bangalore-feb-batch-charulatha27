```python
import pandas as pd
import numpy as np
```

```python
#Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data. (2.5 points)
fb = pd.read_csv('Laliga.csv')
fb.head()
```

```python
#Rearrange dataframe with proper column name
fb  = pd.DataFrame(fb.values[1:], columns=fb.iloc[0])
fb.head()
```

```python
#replace dashes with 0 to make sure you can perform arithmetic operations on the data. (2.5 points)
fb = fb.replace(to_replace="-", value="0")
```

```python
#Print all the teams which have started playing between 1930-1980. (5 points)
print('Teams that started playing between 1930-1980:')
fb[(fb['Debut'] >= '1930') & (fb['Debut'] <= '1980')]['Team']
```

```python
#Print the list of teams which came Top 5 in terms of points (2.5 points)
fb['Points']=pd.to_numeric(fb['Points'])
print('Top 5 Teams in terms of Points')
fb.nlargest(5, 'Points')['Team']
```

```python
###Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. 
###Using the same function, find the team which has maximum and minimum goal difference. (5 points)
###*Goal_diff_count = GoalsFor - GoalsAgainst

def Goal_diff_count(GoalsFor, GoalsAgainst):
    return abs(GoalsFor - GoalsAgainst)

fb['GamesPlayed']=pd.to_numeric(fb['GamesPlayed'])
fb['GoalsFor']=pd.to_numeric(fb['GoalsFor'])
fb['GoalsAgainst']=pd.to_numeric(fb['GoalsAgainst'])

fb['GoalsDiffCount']=Goal_diff_count(fb['GoalsFor'], fb['GoalsAgainst'])
print('Minimum goal diff count team:\n', fb.nsmallest(1,'GoalsDiffCount')['Team'])
print('\n\nMaximum goal diff count team:\n', fb.nlargest(1,'GoalsDiffCount')['Team'])
```

```python
###*5. Create a new column with name “Winning Percent” and append it to the data set (5 points)
###*Percentage of Winning = (GamesWon / GamesPlayed)*100
###*If there are any numerical error, replace it with 0%
###*Print the top 5 teams which has the highest Winning percentage

fb['GamesWon']=pd.to_numeric(fb['GamesWon'])
fb['GamesPlayed']=pd.to_numeric(fb['GamesPlayed'])

fb['Winning Percent']=(fb['GamesWon']/fb['GamesPlayed'])*100

print('Top 5 Winning percentage Teams:\n', fb.nlargest(5,'GoalsDiffCount')['Team'])
```

```python
###*6. Group teams based on their “Best position” and print the sum of their points for all positions (5 points)

###*Eg: Best Position     Points

###*         1                   25000

###*         2                    7000
fb['BestPosition']=pd.to_numeric(fb['BestPosition'])
db = fb.groupby(['BestPosition'], as_index=True).agg({'Points':sum})
print(db)
```
